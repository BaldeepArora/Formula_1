```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
path = 'C:\\Users\\balde\OneDrive\Desktop\Personal Project\Formula_1\datasets'.replace('\\','/')
```

```python
docs = [f for f in os.listdir(path) if os.path.isfile]
```

```python
docs
```

```python
df = {}
for item in docs:
    if item.endswith('.csv'):
        df_name = item.split('.')[0]
        file_path = os.path.join(path,item).replace('\\','/')
        df[df_name] = pd.read_csv(file_path)

        
```

```python
races_circuits = pd.merge(df['races'],df['circuits'],on='circuitId',how='left',suffixes=('_races','_circuits'))
races_results = pd.merge(races_circuits, df['results'], on='raceId', how='left')

# Merge Races+Results with Drivers
races_results_drivers = pd.merge(races_results, df['drivers'], on='driverId', how='left', suffixes=('', '_driver'))

# Merge Races+Results+Drivers with Constructors
races_results_drivers_constructors = pd.merge(races_results_drivers, df['constructors'], on='constructorId', how='left', suffixes=('', '_constructor'))

# # Finally, merge with Status
final_merged_dataset = pd.merge(races_results_drivers_constructors, df['status'], on='statusId', how='left')
```
